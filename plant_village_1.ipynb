```python
dataset = tf.keras.preprocessing.image_dataset_from_directory(data_dir,
                                                              batch_size=32, 
                                                              label_mode='categorical',
                                                              image_size=(224, 224),
                                                              seed=42,
                                                              
                                                              )
```

# Splitting the data into train, test & validation

```python
def split_dataset(dataset, train_split=0.8, test_split=0.1, val_split=0.1):
  ds_size = len(dataset)
  train_size = int(ds_size * train_split)
  val_size = int(ds_size * val_split)


  train_ds = dataset.take(train_size)

  val_ds = dataset.skip(train_size).take(val_size)

  test_ds = dataset.skip(train_size).skip(val_size)

  return train_ds, val_ds, test_ds
```

```python
train_ds, val_ds, test_ds = split_dataset(dataset)
```

```python
train_ds = train_ds.prefetch(buffer_size=tf.data.AUTOTUNE)
val_ds = val_ds.prefetch(buffer_size=tf.data.AUTOTUNE)
test_ds = val_ds.prefetch(buffer_size=tf.data.AUTOTUNE)
```

# Building the model

```python
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', 
                                                  patience=3)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Rescaling(1.0 / 255),
    tf.keras.layers.Conv2D(10, 3, activation='relu', input_shape=(224, 224, 3)),
    tf.keras.layers.Conv2D(10, 3, activation='relu'),
    tf.keras.layers.MaxPooling2D(),
   
    tf.keras.layers.Conv2D(10, 3, activation='relu'),
    tf.keras.layers.Conv2D(10, 3, activation='relu'),
    tf.keras.layers.MaxPooling2D(),

    tf.keras.layers.Conv2D(10, 3, activation='relu'),
    tf.keras.layers.Conv2D(10, 3, activation='relu'),
    tf.keras.layers.MaxPooling2D(),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(len(class_names), activation='softmax'),
])

model.compile(loss='categorical_crossentropy',
              optimizer=tf.keras.optimizers.Adam(),
              metrics=['accuracy'])

history = model.fit(train_ds,
                    epochs=100,
                    steps_per_epoch=len(train_ds),
                    validation_data=val_ds,
                    validation_steps=len(val_ds),
                    callbacks=[early_stopping])
```

```python
import pandas as pd
pd.DataFrame(history.history).plot()
```

```python
model.evaluate(test_ds)
```

```python
for image_test, label_test in test_ds.take(1):
  first_image = image_test[0].numpy().astype('uint8')
  print(label_test[0].numpy())
```

```python
plt.figure(figsize=(20, 15))
for image_test, label_test in test_ds.take(1):
  for i in range(12):
    plt.subplot(6, 2, i+1)
    first_image = image_test[i].numpy().astype('uint8')
    first_label = label_test[i].numpy()

    actual_value = class_names[tf.argmax(first_label)]
  
    prediction = model.predict(tf.expand_dims(first_image, axis=0))

    predicted_value = class_names[tf.argmax(prediction[0])]
  
    if actual_value == predicted_value:
      color = 'green'
    else:
      color = 'red'

    plt.imshow(first_image)
    plt.title(f"Actual: {actual_value}, Predicted: {predicted_value}", color=color)
    plt.axis(False);


```
